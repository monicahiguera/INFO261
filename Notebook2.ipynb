```python
import mysql.connector
import pandas as pd
from IPython.display import display,HTML

db_connection = mysql.connector.connect(user="root",host="localhost",password="root")
cursor = db_connection.cursor()
cursor.execute("USE sakila")
```

```python
print("Contar el número de clientes en la tabla customer")

df = pd.read_sql('SELECT COUNT(*) FROM customer;', con=db_connection)
display(df)
```

```python
print("En la tabla RENTAL, contar cuántos arriendos se realizaron por cada cliente (NB: utilizar la instrucción GROUP BY y COUNT)")

df = pd.read_sql('SELECT COUNT(*) rental_id FROM rental GROUP BY customer_id;', con=db_connection)
display(df)
```

```python
print("Qué hace la siguiente consulta")

df = pd.read_sql('SELECT MAX(total) FROM (SELECT customer_id,count(*) as total FROM rental GROUP BY customer_id) as t1', con=db_connection)
display(df)

print("Muestra en pantalla la mayor cantidad de veces que se ha arrendado, (de lo anterior mostrado)")
```

```python
print("Qué es una sub consulta")

print("RESPUESTA: Una subconsulta es como por ejemplo la consulta anterior, donde se anida una instrucción SELECT dentro de otra.")
```

```python
print("En la tabla PAYMENT mostrar el volumen de negocio realizado por cada miembro del staff (staff_id) en total")

df = pd.read_sql('SELECT COUNT(*) amount FROM payment GROUP BY staff_id;', con=db_connection)
display(df)
```

```python
print("Mostrar el staff_id que ha realizado más volumen de negocio el dia 2005-08-23")

df = pd.read_sql('SELECT MAX(amount) FROM payment WHERE payment_date BETWEEN "2005-08-23 00:00:00" and "2005-08-23 23:59:59";', con=db_connection)
display(df)
```
